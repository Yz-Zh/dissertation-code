### change the format

#### origins: Shanghai

```python
origlat=[]
origlon=[]
for centroid in shanghai["Centroids"]:
    origlon.append(str(centroid).split(' ')[1].split('(')[1])
    origlat.append(str(centroid).split(' ')[2].split(')')[0])
```

```python
origcood1=[]
i=0
while i<214:
    cood=origlon[i]+','+origlat[i]
    origcood1.append(cood)
    i=i+1
print("finish")
```

```python
key='60dfcde343dc69afbb331da01cece742'
urlhead='https://restapi.amap.com/v3/direction/driving?'
```

```python
shanghai_travel_time_car=[[]for i in range(214)]
temp=0
origcount=0
destnumber=25
while origcount<214:
    destcount=0
    while destcount<destnumber:
        url=urlhead+"&origin="+str(origcood1[origcount])+"&destination="+str(destcood1[destcount])+"&key="+key
        result=json.loads(urllib.request.urlopen(url).read())
        temp=temp+1
        print("request NO."+str(temp)+"  "+str(round(temp*100/(214*destnumber),2))+"%")
        #####
        shanghai_travel_time_car[origcount].append(result['route']['paths'][0]['duration'])
        destcount=destcount+1
    origcount=origcount+1
print("finish")
```

#### distances: the schools

```python
dataorigin = gpd.read_file("../shp/school/school.shp")
dataorigin.info()
```

```python
latschool=[]
lonschool=[]
for school in dataorigin["geometry"]:
    lonschool.append(str(school).split(' ')[1].split('(')[1])
    latschool.append(str(school).split(' ')[2].split(')')[0])
```

```python
destcood1=[]
temp=0
while temp<6:
    count=0
    cood=""
    while count<5:
        if count!=4:
            cood=cood+latschool[temp*5+count]+","+lonschool[temp*5+count]+"|"
        else:
            cood=cood+latschool[temp*5+count]+","+lonschool[temp*5+count]
        count=count+1
    destcood1.append(cood)
    temp=temp+1
print("finish")
```

### park time

```python
travel_time_msoa_park_car=[[]for i in range(len(origlat)-1+1)]
origroup=0
while origroup<33:
    destgroup=0
    while destgroup<11:
        origcood=origcood1[origroup]
        destcood=destcood1[destgroup]
        url=urlhead+"&origins="+str(origcood)+"&destinations="+str(destcood)+"&key="+key
        result=json.loads(urllib.request.urlopen(url).read())
        destgroup=destgroup+1
        #####
        if origroup!=32:
            orig=0
            while orig<=20-1:
                dest=0
                while dest<=5-1:
                    # 左索引控制orig;右索引控制dest
                    travel_time_msoa_park_car[(origroup*20+orig)].append(result['rows'][orig]['elements'][dest]['duration']['value'])
                    dest=dest+1
                orig=orig+1
        else:
            orig=0
            while orig<=9-1:
                dest=0
                while dest<=5-1:
                    # 左索引控制orig;右索引控制dest
                    travel_time_msoa_park_car[(origroup*20+orig)].append(result['rows'][orig]['elements'][dest]['duration']['value'])
                    dest=dest+1
                orig=orig+1
    origroup=origroup+1
    print("processed "+"("+str(origroup)+"/"+str(33)+")"+":["+"^"*(origroup)+"_"*(33-origroup)+"] "+str(int(origroup/33*100))+"%")
savecsv=pd.DataFrame(data=travel_time_msoa_park_car)
savecsv.to_csv("test ward travel time car.csv")
print("Finished")
```

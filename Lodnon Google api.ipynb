### import

```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

### load the ward data

```python
wards = gpd.read_file(f"zip://statistical-gis-boundaries-london.zip!statistical-gis-boundaries-london/ESRI/London_Ward.shp")
wards.to_crs("EPSG:4326", inplace = True)
wards["Centroids"] = wards.geometry.centroid
wards
```

### change the format

#### origins: the wards

```python
origlat=[]
origlon=[]
for centroid in wards["Centroids"]:
    origlon.append(str(centroid).split(' ')[1].split('(')[1])
    origlat.append(str(centroid).split(' ')[2].split(')')[0])
```

```python
origcood1=[]
temp=0
while temp<33:
    count=0
    cood=""
    while count<20:
        if (temp*20+count)!=649-1:
            if count!=19:
                cood=cood+origlat[temp*20+count]+","+origlon[temp*20+count]+"|"
            else:
                cood=cood+origlat[temp*20+count]+","+origlon[temp*20+count]
        else:
            cood=cood+origlat[temp*20+count]+","+origlon[temp*20+count]
            break
        count=count+1
    origcood1.append(cood)
    temp=temp+1
print("finish")
```

#### distances: the public space

```python
dataorigin = gpd.read_file("../data/hospital/shp/hospital.shp")
dataorigin
```

```python
latpark=[]
lonpark=[]
for park in dataorigin["geometry"]:
    lonpark.append(str(park).split(' ')[1].split('(')[1])
    latpark.append(str(park).split(' ')[2].split(')')[0])
```

```python
destcood1=[]
temp=0
while temp<5:
    count=0
    cood=""
    while count<5:
        if count!=4:
            cood=cood+latpark[temp*5+count]+","+lonpark[temp*5+count]+"|"
        else:
            cood=cood+latpark[temp*5+count]+","+lonpark[temp*5+count]
        count=count+1
    destcood1.append(cood)
    temp=temp+1
print("finish")
```

#### distances: the schools

```python
dataorigin = gpd.read_file("../shp/school/school.shp")
dataorigin.info()
```

```python
latschool=[]
lonschool=[]
for school in dataorigin["geometry"]:
    lonschool.append(str(school).split(' ')[1].split('(')[1])
    latschool.append(str(school).split(' ')[2].split(')')[0])
```

```python
destcood1=[]
temp=0
while temp<6:
    count=0
    cood=""
    while count<5:
        if count!=4:
            cood=cood+latschool[temp*5+count]+","+lonschool[temp*5+count]+"|"
        else:
            cood=cood+latschool[temp*5+count]+","+lonschool[temp*5+count]
        count=count+1
    destcood1.append(cood)
    temp=temp+1
print("finish")
```

### park time

```python
travel_time_msoa_park_car=[[]for i in range(len(origlat)-1+1)]
origroup=0
while origroup<33:
    destgroup=0
    while destgroup<11:
        origcood=origcood1[origroup]
        destcood=destcood1[destgroup]
        url=urlhead+"&origins="+str(origcood)+"&destinations="+str(destcood)+"&key="+key
        result=json.loads(urllib.request.urlopen(url).read())
        destgroup=destgroup+1
        #####
        if origroup!=32:
            orig=0
            while orig<=20-1:
                dest=0
                while dest<=5-1:
                    # 左索引控制orig;右索引控制dest
                    travel_time_msoa_park_car[(origroup*20+orig)].append(result['rows'][orig]['elements'][dest]['duration']['value'])
                    dest=dest+1
                orig=orig+1
        else:
            orig=0
            while orig<=9-1:
                dest=0
                while dest<=5-1:
                    # 左索引控制orig;右索引控制dest
                    travel_time_msoa_park_car[(origroup*20+orig)].append(result['rows'][orig]['elements'][dest]['duration']['value'])
                    dest=dest+1
                orig=orig+1
    origroup=origroup+1
    print("processed "+"("+str(origroup)+"/"+str(33)+")"+":["+"^"*(origroup)+"_"*(33-origroup)+"] "+str(int(origroup/33*100))+"%")
savecsv=pd.DataFrame(data=travel_time_msoa_park_car)
savecsv.to_csv("test ward travel time car.csv")
print("Finished")
```
